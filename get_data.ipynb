```python
# def get(df):
#     res = df[df['cause'].isin(list_of_cause) & df['measure'].isin(list_of_measuers) & df['metric'].isin(list_of_metrics) & df['location'].isin(list_of_locations)]
#     return res

def fix_long_name(df):
    df.measure = df.measure.str.replace('YLLs (Years of Life Lost)', 'YLLs', regex=False).str.replace('DALYs (Disability-Adjusted Life Years)', 'DALYs', regex=False).str.replace('YLDs (Years Lived with Disability)', 'YLDs', regex=False)
    return df

def save(df, file_name):
    # sepeate all ages from age ranged rows
    all_ages = df[df.age=="All Ages"]
    all_ages.to_csv(f"{root_path}/res/all/{file_name}_all.csv", index=False)

    for i in ['YLLs', "DALYs", "YLDs", "Deaths"]:
        to_save  = df[(df.age!="All Ages") & (df.measure==i)]
        to_save.to_csv(f"{root_path}/res/{i}/{file_name}_{i}.csv", index=False)
    # seperate the risk for ages
```

```python
for name in csv_names:
    print('name', name)
    df = pd.read_csv(name)
    df = fix_long_name(df)
    # df = get(df)
    file_name = name.split('/')[-1].split('.')[0]
    save(df, file_name)
```

```python
def concat_csv(type_name, csvs_list):
    list_of_df = []
    for l in csvs_list:
        print('read l', l)
        list_of_df.append(pd.read_csv(l))

    print('concat')
    df =pd.concat(list_of_df)
    print('save')
    df.to_csv(f"/home/amir/big_storage/datasets/IHME-GDB-Risk-{type_name}.csv", index=False)
```

```python

dirs = os.listdir(f"{root_path}/res")
for folder in dirs:
    file_list = glob.glob(f"{root_path}/res/{folder}/*.csv")
    concat_csv(folder, file_list)
    
```
